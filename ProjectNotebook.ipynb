```python
# Import all trailers from pickle
pickle_off = open("pickles/trailersdataofficial.pickle","rb")
trailersdata = pickle.load(pickle_off)
```

```python
# Calculating total trailer time in seconds
for vid in trailersdata:
    if "M" in vid['duration']:
        if "S" in vid['duration']:
            timearr = (vid['duration'].split('PT')[1].replace('S','').split('M'))
            totalduration = (float(timearr[0])*60)+float(timearr[1])
            vid['durationseconds'] = totalduration
        else:
            totalduration = float(vid['duration'].split('PT')[1].split('M')[0])
            vid['durationseconds'] = totalduration*60
    else:
        totalduration = float(vid['duration'].split('PT')[1].split('S')[0])
        vid['durationseconds'] = totalduration
```

```python
# Getting word counts by trailer
for vid in trailersdata:
    text = TextBlob(vid['transcripttext'])
    vid['wordcount'] = len(text.words)
```

```python
# Get a list of all possible movie genres
all_genres=[]
for vid in trailersdata:
    try:
        for genre in vid['imdb_genres']:
            if genre not in all_genres:
                all_genres.append(genre)
    except:
        pass
```

```python
# Splitting trailers into genres
trailersByGenre = {}
for index, genre in enumerate(all_genres):
    genre_list = []
    for vid in trailersdata:
        try:
            if genre in vid['imdb_genres']:
                genre_list.append(vid)
        except:
            pass
    genrename = genre + "_genre"
    trailersByGenre[genrename] = genre_list
```

```python
# Getting array enumerating number of movies for each genre
array_of_genres = []
for genre, arr_of_trailers in trailersByGenre.items():
    array_of_genres.append((len(arr_of_trailers), genre))
array_of_genres.sort(key=lambda x: x[0], reverse=True)
array_of_genres    
```

```python
# Get avg stats of any dictionary that has keys where values are list of trailers (will use for imdb rating dictionary as well)
def getstats(dictionary):
    df = pd.DataFrame(columns=["type","avg_words","speechtime","duration"])
    for key in dictionary:
        counts = []
        speechtime = []
        lengths = []
        for vid in dictionary[key]:
            counts.append(vid['wordcount'])
            speechtime.append(vid['speechduration']/vid['durationseconds'])
            lengths.append(vid['durationseconds'])

        genre = key
        avg_words = round(np.average(counts),3)
        speechtime = round(np.average(speechtime),3)
        duration = round(np.average(lengths),3)
        df.loc[len(df)] = [genre,avg_words,speechtime,duration] 
    return df
```

# Clustering Movie Trailers By Sentiment

Need to normalize all trailers to same length

```python
# print out avg duration of trailers
# average duration of a trailer is around 2 minutes so I made that the standard in code below
durations = []
for vid in trailersdata:
    durations.append(vid['durationseconds'])
np.average(durations)
```

```python
# Function for time warping any trailer into a standardized set of points in order to compare them later when clustering them
def timewarp(trailer, totaltime, intervaltime):
    points = []
    duration = trailer['durationseconds']
    warper = (duration/totaltime)

    timeinterval = totaltime/intervaltime
    intervallist = list(np.linspace(start=0.0, stop=totaltime, num=(timeinterval+1.0)))
    bucketlist = []
    for first, second in zip(intervallist, intervallist[1:]):
        bucketlist.append((first, second))
    for bucket in bucketlist:
        bucket_sentiments = []
        for sentence in trailer['sentences']:
            if (bucket[0] < round(float(sentence['starttime'])/warper,2) <= bucket[1]):
                bucket_sentiments.append(sentence['sentimentVader'])
        if bucket[0] == 0.0:
            points.append((bucket[0],0.0))
        elif bucket_sentiments:
            points.append((bucket[0],np.average(bucket_sentiments)))
        else:
#             points.append((bucket[0],np.nan))    
            points.append((bucket[0],0.0))    
    points.append((totaltime,points[-1][1]))
    return points
```

```python
# Add timewarped points to trailers in order to cluster late, normalizing at 2 min trailer and 5 second intervals
for trailer in trailersdata:
    warpedpoints = timewarp(trailer, 120, 15)
    trailer['normalizedsentimentpoints'] = warpedpoints
```

```python
# Plot a trailer's timewarped points of sentiment
def plotwarpedtrailer(index):
    starttimes = [x[0] for x in trailersdata[index]['normalizedsentimentpoints']]
    sentiment = [x[1] for x in trailersdata[index]['normalizedsentimentpoints']]
    plt.figure(figsize=(10,5))
    plt.plot(starttimes,sentiment, '-rD')
    plt.title(trailersdata[index]['imdb_title'])
    plt.xlabel("Seconds")
    plt.ylabel("Sentiment")
    plt.xlim((0, 120))
    plt.ylim((-.9,.9))
    print(trailersdata[index]['youtube_url'])
    return plt
```

```python
# Write function to plot avg trailer sentiment for each genre
def plotgenresentiment(genre):
    sentiments = []
    for vid in trailersByGenre[genre]:
        sentiments.append([x[1] for x in vid['normalizedsentimentpoints']])
    starttimes = [0.0, 15.0, 30.0, 45.0, 60.0, 75.0, 90.0, 105.0, 120]
    avgsentiments = [sum(e)/len(e) for e in zip(*sentiments)]
    
    plt.figure(figsize=(10,5))
    plt.plot(starttimes,avgsentiments, '-rD')
    plt.title(genre + " Avg Sentiment")
    plt.xlabel("Seconds")
    plt.ylabel("Sentiment")
    plt.xlim((0, 120))
    plt.ylim((-.4,.4))
    return plt   
```
